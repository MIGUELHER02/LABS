```python
import pandas as pd
from matplotlib import pyplot as plt
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("Problema 1.1")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(["y", "z"])
plt.show()
```

```python
# Analisis de datos countries
```

```python
data = pd.read_csv("countries.csv")
```

```python
data
```

```python
guatemala = data[data.country == "Guatemala"]
```

```python
guatemala
```

```python
honduras = data[data.country == "Honduras"]
```

```python
honduras
```

```python
plt.plot(guatemala.year, guatemala.population/10**6)
plt.plot(honduras.year, honduras.population/10**6)
plt.legend(["Guatemala", "Honduras"])
plt.title("Poblacion Honduras y Guatemala")
plt.xlabel("Año")
plt.ylabel("Millones de Habitantes")
plt.show()
```
